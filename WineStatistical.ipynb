"Hồi quy tuyến tính" là một **phương pháp thống kê để hồi quy dữ liệu** với biến phụ thuộc có giá trị liên tục trong khi các biến độc lập có thể có một trong hai giá trị liên tục hoặc là giá trị phân loại. Nói cách khác "Hồi quy tuyến tính" là một phương pháp để dự đoán biến phụ thuộc (Y) dựa trên giá trị của biến độc lập (X). Nó có thể được sử dụng cho các trường hợp chúng ta muốn dự đoán một số lượng liên tục. Ví dụ, 
- dự đoán giao thông ở một cửa hàng bán lẻ, 
- dự đoán thời gian người dùng dừng lại một trang nào đó hoặc số trang đã truy cập vào một website nào đó v.v...
- dự đoán các biểu đồ tăng trưởng 

```
import pandas as pd
import numpy as np
 
wine = pd.read_csv("./data/winequality-red.csv", sep=";")
wine.head()
from sklearn import linear_model
clf = linear_model.LinearRegression()
display(wine)
# Sử dụng "density (mật độ)" làm biến giải thích
X = wine.loc[:, ['density']].to_numpy()
 
# Sử dụng "alcohol (Số độ cồn)" làm biến mục đích
Y = wine['alcohol'].to_numpy()
 
# Tạo model suy đoán
clf.fit(X, Y)
 
# Hệ số hồi quy
print(clf.coef_)
 
# Sai số
print(clf.intercept_)
 
# Score
print(clf.score(X, Y))
```

```
# sử dụng package matplotlib
import matplotlib.pyplot as plt
 
# Biểu diễn sự phân bố tập dữ liệu input
# c: color
plt.scatter(X, Y, c='b')
 
# Đường thẳng hồi quy
plt.plot(X, clf.predict(X))
plt.show()
```
